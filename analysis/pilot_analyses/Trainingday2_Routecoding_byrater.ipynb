```python
data = d.groupby(['subid', 'rep', 'code1']).count().reset_index()
data = pd.pivot_table(data, index=['subid', 'rep'], 
                      columns='code1', values='env', fill_value=0).reset_index()
print data.shape
data.head()
```

```python
data_counts = d.groupby(['subid', 'rep']).count().reset_index()
print data_counts.shape
data_counts.head()
```

```python
len(data.subid.unique())
```

```python
data['total_trials'] = data_counts.env
data.head()
```

```python
d1 = data.loc[:,codes[:4]]
d2 = data.loc[:,'total_trials']
proportions = d1.divide(d2, axis='rows')
proportions.head()
```

```python
proportions[['subid', 'rep']] = data[['subid', 'rep']]
proportions.head()
```

```python
proportions = pd.melt(proportions, id_vars=['subid', 'rep'])
proportions.head()
```

```python
data = proportions.loc[(proportions.rep == 2)]

sns.boxplot(x='code1', y='value', hue='group',
            data=data, palette=['white'])
sns.swarmplot(x='code1', y='value', hue='group', split=True,
              palette=['dodgerblue', 'orange'],size=7,linewidth=1,
               data=data)

plt.ylabel('Proportion habit trials')
```
