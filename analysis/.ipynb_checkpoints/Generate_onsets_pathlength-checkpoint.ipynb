### Calculate path lengths:

```python
df_in.rep.value_counts() # 3 reps for sst20, env 456
```

```python
# Get just the rows where position in space is logged
dl = df_in.loc[df_in.command == 'VROBJECT_POS']
dl.head()
```

```python
d_pathlen = dp_piv.copy()
d_pathlen['path_length'] = 0

for trial in dp_piv.itertuples():

    subid = trial.subid
    scan = trial.scan
    env = trial.env

    # get sub data for this run/environment
    data_sub = dl.loc[(dl.subid == subid) & (dl.scan == scan) & (dl.env == env)]

    # prune timepoints where position logged outside of actual navigation 
    # (usually some at end/beginning of trials, e.g., at y=0)
    data_sub = data_sub.loc[(data_sub.time_adj < trial.ARRIVED) & (data_sub.time_adj > trial.NAVIGATE)]

    # calculate path length (sum of distances between points)
    points = zip(data_sub.x, data_sub.y)
    point_distance = np.diff(points, axis=0) # diff between steps
    #hypotenuse between sequential steps, summed over samples:
    distance = sum(np.sqrt((point_distance ** 2).sum(axis=1))) 
    
    # add to dataframe
    d_pathlen.loc[(d_pathlen.subid == subid) & 
                  (d_pathlen.scan == scan) & 
                  (d_pathlen.env == env), 'path_length'] = distance

d_pathlen.to_csv('/Volumes/group/awagner/sgagnon/SST/nav_data/analysis/fmri_path_length.csv')
```

```python
## Trim off beginning of run 5 (to start with env 3, rather than re-run of env5)
starting_trs = 82 # how many trs in raw ts

# based on onset, starting TR:
onset = 83.851 # onset, adjusted for the dropped 4 TRs
print 'Trimmed ts onset TR: ' + str(np.ceil(onset / 2.))
print 'Raw ts onset TR: ' + str(np.ceil((onset + 8) / 2.))

tr_to_drop = (np.ceil((onset + 8) / 2.) - 1) - 4 # keep 1 before for buffer + 
                                                 # keep 4 more to be dropped during preproc
print 'Raw ts TR to drop: ' + str(tr_to_drop)
print 'Updated raw ts will be this length: ' + str(starting_trs - tr_to_drop)

# new onset (after dropping 4 TRs during preproc):
print 'New onset: ' + str((onset) - (tr_to_drop)*2.)
```

### Onsets by run type

```python
# model runs by task (habit/probe), and shock where relevant
output_filename = 'navcond_byruntype.csv'
output_dir = '/Volumes/group/awagner/sgagnon/SST/data'

for sub, group in zip(subj_info.subid, subj_info.group):

    print sub
    
    dsub = dp_piv.loc[(dp_piv.subid == sub)]
    
    # iterate through the events
    ddesign = pd.DataFrame()
    for event, dur in zip(['ORIENT', 'ASSIGNED', 'NAVIGATE', 'ARRIVED'], [11, 0, 0, 0]):
        
        # get nav duration if that event, otherwise set boxcar/impulse
        if event == 'NAVIGATE':
            dur = dsub.nav_dur

        ddesign_ev = pd.DataFrame({'run': dsub.scan, 
                                   'condition': event + '_' + dsub.run_type,
                                   'onset': dsub[event],
                                   'duration': dur,
                                   'value': 1})
        ddesign = ddesign.append(ddesign_ev, ignore_index=True)
        
    # add in shocks if relevant
    if group == 'stress':
        for event, dur in zip(['SHOCK'], [0]):

            ddesign_ev = pd.DataFrame({'run': dsub.loc[~dsub.SHOCK.isnull(), 'scan'], 
                                       'condition': event,
                                       'onset': dsub.loc[~dsub.SHOCK.isnull(), 'SHOCK'],
                                       'duration': dur,
                                       'value': 1})
            ddesign = ddesign.append(ddesign_ev, ignore_index=True)
    
    if int(sub) < 10:
        sub_output_dir = op.join(output_dir, 'sst0' + str(sub), 'design')
    else:
        sub_output_dir = op.join(output_dir, 'sst' + str(sub), 'design')
        
    print sub_output_dir
    
    if not os.path.exists(sub_output_dir):
        os.makedirs(sub_output_dir)
    
    ddesign.to_csv(op.join(sub_output_dir, output_filename), index=False)
```
