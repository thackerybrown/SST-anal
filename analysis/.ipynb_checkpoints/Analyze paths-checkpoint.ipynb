# Analyze navigation paths

## Quick background

In this study, participants learn a familiar route around 12 different virtual environments over 2 days of training. On the third day, participants complete two tasks inside an MR scanner while we measure their neural activity. 

During the **familiar route/habit task** participants are placed along the well-learned route, given a 360 orientation of their location, and then given the instructions to navigate the route they had learned (~9s of planning time). Navigation was self-paced. 

Following this task, participants completed the **probe/shortcut task**. Here, they were placed along the well-learned route, given a 360 orientation of their location, and then given the instructions to navigate to a goal item within the environment, thinking about *where* this item is located in the environment, and *how* they will get to this item as quickly as possible, using whatever strategy they want (~9s of planning time).

Finally, participants were divided into two groups on the scan day (control, stress), so we could examine **the effect of stress** on the ability to plan and flexibly navigate (vs. rely on more "automatic" responses).

![alt text](http://i.imgur.com/lI1cS2T.jpg "Title")

## Goal

One of our goals was to measure neural signal during **planning**, and to examine whether increased activity in the **hippocampus** (thought to store a "cognitive map", and enable retrieval of memories) and **lateral frontopolar cortex** (critical for planning, particularly thinking about subgoals while maintaining the longer-term goal in mind) predicted the **efficiency of navigation** during probe trials (i.e., taking flexible shortcuts to the goal location). In addition to our coding of trials as "shortcuts", "backtrack", "familiar route" or some combination of strategies/wandering ("other"), we can also look at measures derived from navigation behavior, such as number of pauses. We'll calculate that in this notebook.

## Read in route info / calculate path length, speed

Here, we'll begin by reading in the dataframe `group_fmri_onsets.h5` that we generated in [SST_behav_fmri_generateData.ipynb](https://github.com/sgagnon/SST/blob/master/analysis/SST_behav_fmri_generateData.ipynb) by combining the log files for each trial intelligently (e.g., updating scan number when runs crashed, merging environment information based on subject-specific counterbalancing) and down-sampling navigation data to ~2Hz. Consequently this dataframe has may rows (~400k) logging avatar position/heading direction twice every two seconds, in addition to button presses, event onsets, etc.

#### Structure of data
This `group_fmri_onsets.h5` dataframe contains all the navigation information for our participants, including the avatar position (`x,y,z` coordinates) and heading direction, event flags marking the phases of the trial ("ORIENT": initial orientation to the environment, "ASSIGNED": planning phase, "NAVIGATE": navigation phase, "ARRIVED": goal arrival), keyboard events, the task (`type`; familiar route/habit or probe/shortcut), the environment label (`env`), the MRI scan number (`scan`), and the time relative to the onset of the scan (`time_adj`; adjusted for 4 TRs/8s that is discarded to allow for stabilization of the MR signal; these our our event onsets used to label the fMRI data).

#### Preprocessing
Before working with the dataframe, we need to do some minor preprocessing to (1) remove some avatar positions logged between trials (when the screen is black with a fixation, but the avatar is placed at the 0,0 coordinate), and (2) flip the labeling of 2 scans which were flipped during initial data frame generation.

#### Dataframes:
Then, we'll create two main dataframes. `dp_piv` will contain a row for each unique trial (~1.5k rows: 38 participants * 36 trials), and will be critical for mapping trial information to planning phase neural activity; we can add trial-specific calculations (e.g., navigation duration, average speed, number of pauses) to this dataframe. We'll also create a second dataframe, `dmetrics`, logging various continuous navigation metrics for each participant (e.g., distance traveled between successive samples, speed); this  dataframe will be used for plotting routes, and calculating other trial-wise estimates (e.g., number of pauses). 

```python
# Read in file (generated in SST_behav_fmri_generateData)
filename = '/Volumes/group/awagner/sgagnon/SST/nav_data/group_fmri_onsets.h5'
df_in = pd.read_hdf(filename, 'df')
df_in.shape

########################################################
# Quick preprocessing
########################################################

# Remove some between trial stuff (scan has incremented, but still collecting data)
df_in = df_in.loc[df_in.time_adj > 0]

# Flip scans for sub w/crash, due to glob error; 
# note, added sort to glob, so this would be fixed if generated data again
print 'Flipping scans 3 and 4 for sst54, due to crash/glob error...'
df_in.loc[(df_in.subid == '54') & 
          (df_in.run == 3) & (df_in.command == 'ORIENT')]

df_in.loc[(df_in.subid == '54') & (df_in.run == 3) & 
          (df_in.env.isin(['env5', 'env3'])), 'scan'] = 4
df_in.loc[(df_in.subid == '54') & (df_in.run == 3) & 
          (df_in.env.isin(['env4'])), 'scan'] = 3

df_in.loc[(df_in.subid == '54') & 
          (df_in.run == 3) & (df_in.command == 'ORIENT')]

########################################################
# Pull out onsets for events of interest
########################################################

print 'Pulling out onsets of interest...'
# Create pivot table of main events of interest/onsets (rel to scan start - 4 TRs: for onsets)
dp = df_in.loc[df_in.command.isin(['ORIENT', 'ASSIGNED', 'NAVIGATE', 'ARRIVED', 'SHOCK'])]
dp.head()

dp_piv = dp.pivot_table(index=['subid', 'scan', 'run', 'rep', 'env'], 
                        columns='command', values='time_adj').reset_index()

# Get nav duration
dp_piv['nav_dur'] = dp_piv.ARRIVED - dp_piv.NAVIGATE

# assign run type
dp_piv['type'] = 'habit'
dp_piv.loc[dp_piv.run > 4, 'type'] = 'shortcut'

########################################################
## Calculate nav matrics: 
## path lengths, speed, etc.
########################################################

print 'Calculating metrics for each trial...'
# Get just the rows where position in space is logged
dl = df_in.loc[df_in.command == 'VROBJECT_POS']

# init df to store distance traveled, speed, etc. between successive points
dmetrics = pd.DataFrame() 

# Iterate through trials, and calculate metrics 
dp_piv['path_length'] = 0

for trial in dp_piv.itertuples():

    subid = trial.subid
    scan = trial.scan
    env = trial.env

    # get sub data for this run/environment
    data_sub = dl.loc[(dl.subid == subid) & 
                      (dl.scan == scan) & 
                      (dl.env == env)]

    # prune timepoints where position logged outside of actual navigation 
    # (usually some at end/beginning of trials, e.g., at y=0)
    data_sub = data_sub.loc[(data_sub.time_adj < trial.ARRIVED) & 
                            (data_sub.time_adj > trial.NAVIGATE)]

    # Reset index
    data_sub = data_sub.reset_index()
    
    # Get distance traveled in x, y dimensions between samples
    points = zip(data_sub.x, data_sub.y)
    point_distance = np.diff(points, axis=0) # diff between samples
    
    # Get hypotenuse between sequential steps (i.e., actual distance traveled)
    traveled_distance = np.sqrt((point_distance ** 2).sum(axis=1))
    
    # Calculate speed
    time_diff = np.diff(data_sub.time, axis=0)
    speed = traveled_distance/time_diff
    
    # Sum over samples (sum of distances between points) to get total path length
    path_len = sum(traveled_distance)
    
    # add to dataframe of trials
    dp_piv.loc[(dp_piv.subid == subid) & 
               (dp_piv.scan == scan) & 
               (dp_piv.env == env), 'path_length'] = path_len
    
    # Store metrics along route
    data_sub = data_sub.loc[1:,['subid', 'run', 'rep', 'scan', 'time_adj', 'trial', 'type', 'env', 'x', 'y']]
    data_sub['dist'] = traveled_distance
    data_sub['time_diff'] = time_diff
    data_sub['speed'] = speed
    
    dmetrics = dmetrics.append(data_sub, ignore_index=True)
    
# Remove any super fast time points (due to staggered/fast resampling)
dmetrics = dmetrics.loc[dmetrics.speed < 6].reset_index()
```

### Plot paths (color coded by speed) & flag pauses

We'll use `ipywidgets` to flexibly plot the paths for each participant, **color coding/scaling the trajectory by participants speed** (red/larger circles = slower speed). We'll also add in an **arrow** to mark the start of the route, and underlay each route with the environment layout (gray dots for buildings, and colored stars for items within the environment).

To look at the **speed as a function of time**, we'll also plot that effect below the map, flagging "extreme" slow downs (-3 SD from the mean), and counting the number of pauses. 

#### Counting pauses:
Speed for each trial was used to calculate the number of pauses along that route (using our function `calculate_spikes()`); a “pause” was defined as a downward spike in speed. Specifically, for each trial we z-scored speed across all samples, and then identified samples with speed below 3 SD of the mean as “slow” samples. Contiguous “slow” samples were defined as one pause. In addition, if the route started out as “slow” (i.e., was not a downward deflection), this was defined as a pause. For familiar route trials, participants were instructed to head into the finish line at a 90 degree angle, and consequently often paused to align themselves before heading into the finish line; to avoid including this pause in our count, we removed the last four samples (~2s) of each trial. 

```python
def plot_sub(subj, env, test_trial, rep):
    data = dmetrics.loc[(dmetrics['type'] == test_trial) &
                        (dmetrics.rep == rep) & 
                        (dmetrics.env == env) &
                        (dmetrics.subid == subj)]
    
    plot_path_z(data=data, proj=proj, dirs=dirs, z_dim='speed')
    plt.show()
    
    # add plot of speed over time
    z_speed = pd.Series(sp.stats.zscore(data.speed))
    
    plt.plot(z_speed)

    # overlay spike count + deviations from mean
    spike_count, mask = calculate_spikes(data.speed, test_trial)
    
    plt.plot(z_speed.index[mask], z_speed[mask], 'ro', alpha=.5)
    plt.xlabel('Time (s)'); plt.ylabel('z-scored speed')
    plt.hlines(-3, 0, len(z_speed), linestyles='dashed', colors='gray')
    plt.text(0.1, 0.95, '  # of pauses: ' + str(spike_count), 
            fontsize=14, verticalalignment='top')

subj_opts = ipywidgets.Select(options=list(dmetrics.subid.unique()))
env_opts = ipywidgets.ToggleButtons(options=list(np.sort(dmetrics.env.unique())))
test_opts = ipywidgets.ToggleButtons(options=list(['habit', 'shortcut']))
rep_opts = ipywidgets.IntSlider(min=1, max=2)

w = ipywidgets.interact(plot_sub, 
                        subj=subj_opts, 
                        env=env_opts, 
                        test_trial=test_opts, 
                        rep=rep_opts)
```

```python
dmetrics['trial_bin'] = dmetrics.groupby(['subid', 'run', 'rep', 'scan', 
                                               'trial', 'type', 'env'])['time_adj'].transform(
    lambda x: pd.qcut(x, 10, labels=range(1,10+1)))

# just to double check that equal #s per bin:
# dmetrics.groupby(['subid', 'run', 'rep', 'scan', 'trial', 'type', 'env', 'trial_bin']).count()

# calculate mean/sd speed for each trial
zscore = lambda x: (x - x.mean()) / x.std()

dmetrics['speed_z'] = dmetrics.groupby(['subid', 'type', 'env', 'rep'])['speed'].transform(zscore)
# dmetrics['slow'] = 
dmetrics['speed_slow'] = dmetrics.speed_z < -3
dmetrics.head()
```

### Calculate the number of stops/pauses for each trial

Notes:
- Some stops are due to slow turning around corners
- Stops at the end of familiar route trials are due to re-alignment before going into finish line (toss last 4 samples from familiar route trials)

```python
dp_piv['pauses'] = 0

for trial in dp_piv.itertuples():
        
    # calc speed
    spikes,_ = calculate_spikes(dmetrics.loc[(dmetrics['type'] == trial.type) &
                                             (dmetrics.rep == trial.rep) & 
                                             (dmetrics.env == trial.env) &
                                             (dmetrics.subid == trial.subid)].speed, 
                                trial.type, verbose=False)
    
    dp_piv.loc[(dp_piv['type'] == trial.type) &
               (dp_piv.rep == trial.rep) & 
               (dp_piv.env == trial.env) &
               (dp_piv.subid == trial.subid), 'pauses'] = spikes
```
