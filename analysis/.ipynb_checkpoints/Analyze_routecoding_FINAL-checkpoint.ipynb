Goal is to analyze route coding from:
    - Training day 2 (habit task)
    - Test day (habit task)
    - Test day (probe task, 2 reps)

```python
# Remove subjs w/bad coding
remove_subjs = ['sst01', 'sst02', 'sst39']
d = d.loc[~d.subid.isin(remove_subjs)]

print d[d.final_code.isnull()]
```

### Update final_code with final_code2 (convert some "other" to shortcut)

```python
d.loc[~d.final_code2.isnull(), 'final_code'] = d.loc[~d.final_code2.isnull(), 'final_code2']
d.head()
```

```python
d.final_code.value_counts()
```

## Get trial counts/type for subjects/days/tasks:

```python
data = d.groupby(['subid', 'day', 'run_type', 'rep', 'final_code']).count().reset_index()
data = pd.pivot_table(data, index=['subid', 'day', 'run_type', 'rep'], 
                      columns='final_code', values='env', fill_value=0).reset_index()
print data.shape
data.head()
```

```python
data_counts = d.groupby(['subid', 'day', 'run_type', 'rep']).count().reset_index()
print data_counts.shape
data_counts.head()
```

```python
len(data.subid.unique())
```

```python
data['total_trials'] = data_counts.env
data.head()
```

```python
data.loc[(data.run_type == 'shortcut') & 
         (data.other >4)]
```

### Calculate proportions

```python
d1 = data.loc[:, codes[:4]]
d2 = data.loc[:,'total_trials']
proportions = d1.divide(d2, axis='rows')

# combine w/dataframe
proportions[['subid', 'day', 'run_type', 'rep']] = data[['subid', 'day', 'run_type', 'rep']]
proportions.head()

# convert from wide (cols are codes) to long (more rows per subj)
proportions = pd.melt(proportions, id_vars=['subid', 'day', 'run_type', 'rep'])
proportions.head()
```

```python
# remove bad/missing subjs
print group_info.loc[group_info.remove == 1]

proportions = proportions.merge(group_info)
proportions = proportions.loc[(proportions.remove!=1)]
```

```python
# any subjs not following habit instructions > 50% time?
proportions.loc[(proportions.run_type == 'habit') & 
                (proportions.final_code == 'habit') & 
                (proportions.value < .5)]
```

# Visualize data:

## Habit task run (day 3):

```python
data = proportions.loc[(proportions.run_type =='habit') & (proportions.day == 'day3')]

sns.factorplot(x='final_code', y='value', 
               ci=68, units='subid', data=data, aspect=2)
sns.stripplot(x='final_code', y='value', jitter=True, hue='group', 
              palette=['dodgerblue', 'orange'],size=7,linewidth=1,
               data=data)

plt.ylabel('Proportion habit trials')
```

```python
proportions.loc[(proportions.run_type=='habit') & 
                (proportions.final_code == 'habit'), 'value'].hist()
```

```python
x_order = ['shortcut', 'backtrack', 'other', 'habit']

p = sns.factorplot(x='final_code', y='value',
                   order=x_order, aspect=1.2,
                   ci=68, units='subid', data=proportions.loc[(proportions.day == 'day3') &
                                                              (proportions.run_type == 'habit')], 
                   kind='bar')

p.set_xlabels('')
p.set_ylabels('proportion of trials')
sns.despine(trim=True)
```

```python
data = proportions.loc[proportions.run_type == 'habit']
data = data.merge(group_info)

p = sns.factorplot(x='final_code', y='value', hue='group', col='day',
                   row='rep',
                   ci=68, units='subid', order=x_order, kind='bar',
                   aspect=1.5, dodge=.1, palette=['dodgerblue', 'orange'],
                   data=data)
p.set_xlabels('')
plt.ylim(0,1)
p.set_ylabels('proportion of trials')
sns.despine(trim=True)

plt.savefig('/Volumes/group/awagner/sgagnon/SST/results/figures/route_coding/habit_choices_bygroup.png', dpi=300)
```
