# SST: Analysis for Item Test

###Procedure

**Session 1**: 

1. Watch 4 videos of each town (2/town in a row)
2. <span style="color:red"> Item test (run 1)</span>
3. Navigation with arrows (2/town)
4. <span style="color:red">Item test (run 2)</span>

**Session 2**:

1. Watch 1 video of each town (2 for subjects 1-4)
2. <span style="color:red">Item test (run 3)</span>
3. Navigation with arrows (2/town)
4. Free navigation without arrows (2/town)
5. <span style="color:red">Item test (run 4)</span>

**Session 3**:

1. Test trials, navigate to item in each environment (2 cycles)
2. Questionnaire about strategies used, etc.
3. <span style="color:red">Item test (run 5)</span>
4. Map drawing
5. Spatial preference questionnaire

### Item test: 
Present subjects with image from town, and asked to recall the 3 items located within that town and to type the names of the items. If they didn't remember an item, they responded with a "?".


Sample image from town:

![env3](./images/env3.jpg)


## Import packages

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os.path as op
import numpy as np

# for plotting
sns.set(style='whitegrid', context='poster')
%matplotlib inline
```

### Define color palette

```python
palette = {'face': 'orange',
           'animal': 'dodgerblue',
           'object': 'lightseagreen',
           'food':  'mediumseagreen',
           'fruitveg':  'mediumseagreen',
           'tool': 'mediumpurple'}
```

## Setup paths

```python
dirs = dict()
dirs['basedir'] = op.join(op.expanduser('~'), 'Experiments/SST')
dirs['datadir'] = op.join(dirs['basedir'], 'data')
dirs['analydir'] = op.join(dirs['basedir'], 'analysis')
dirs['subj_info_file'] = op.join(dirs['datadir'], 'subj_info.csv')
dirs['object_cat_file'] = op.join(dirs['datadir'], 'object_cat.csv')
```

```python
subj_info = pd.read_csv(dirs['subj_info_file'])
subj_info = subj_info[pd.isnull(subj_info.remove)]
subj_info[['subid']] = subj_info.subid.astype(str)

subj_info
```

## Load in data

```python
dt = pd.DataFrame() 
test = True
questionnaires_post = True

for subid in subj_info.subid:
    print subid

    if test:
        # add test file
        test_file = op.join(dirs['datadir'], 'itemTest_coded', subid + '_behav_itemTest_coded.csv')
        d2 = pd.read_csv(test_file)
        d2['subid'] = subid
        dt = dt.append(d2, ignore_index=True)
        
# Compiled group data
if questionnaires_post:
    q_file = op.join(dirs['basedir'], 'data/Quest/Questionnaires_group.csv')
    dq_post = pd.read_csv(q_file, index_col=0)
    dq_post['subid'] = dq_post.subid.astype(str)
```

```python
dt.head()
```

```python
dt = dt.merge(subj_info)
```

#### Fix any typos in itemType column

```python
dt.itemType.value_counts()
```

```python
dt.loc[dt.itemType == 'tool ', 'itemType'] = 'tool'
dt.loc[dt.itemType == 'frtuiveg', 'itemType'] = 'fruitveg'
dt.itemType.value_counts()
```

### Any missing data?

- Missing run 4 for subj 9 (might have been overwritten)
- Missing first 2 runs for subj 16; also likely overwritten

```python
# Any missing runs?
acc_by_run = dt.pivot_table(index=['subid', 'group'], columns='run', values='acc').reset_index()
acc_by_run.loc[acc_by_run.isnull().any(axis=1)]
```

##### Impute values for subj 9 run 4, since they had perfect performance on runs 3 and 5

```python
# Impute values for subj 9 missing run 4 as 100%
sst9_run4 = dt.loc[(dt.subid == '9') & (dt.run == 3)]
sst9_run4.loc[:,'run'] = 4

print dt.shape
dt = dt.append(sst9_run4, ignore_index=True).reset_index()
print dt.shape

# confirm that correct stuff added on
print 7128 - 7092
print sst9_run4.shape[0]
```

```python
run4_errors = dt.loc[(dt.run == 4) & (dt.acc == 0)]
run4_errors['error_type'] = run4_errors.resp
run4_errors.loc[run4_errors.error_type.isin(['?', '/']), 'error_type'] = 'omission'
run4_errors.loc[run4_errors.error_type != 'omission', 'error_type'] = 'other'
# errors for faces
run4_errors = pd.DataFrame(run4_errors.groupby(['subid', 'itemType', 'error_type']).trial.count()).reset_index()
run4_errors.head()

run4_errors = run4_errors.pivot_table(index=['subid'], 
                                      columns=['itemType', 'error_type'], 
                                      values='trial', fill_value=0)
run4_errors = run4_errors.unstack().reset_index()
```

```python
dat = run4_errors.groupby(['subid', 'itemType', 'error_type']).sum().reset_index()
dat['num_errors'] = dat[0]
sns.factorplot(x='error_type', y='num_errors', col='itemType', hue='subid', 
               dodge=.2,
               data=dat, legend=False)
```
